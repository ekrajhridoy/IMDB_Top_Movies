```python

with open('movies.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python

soup = BeautifulSoup(html,'lxml')
```

```python

containers = soup.find_all('div',{'class':'ipc-poster-card ipc-poster-card--base ipc-poster-card--media-radius ipc-poster-card--dynamic-width sc-d017d8-0 lkttPu gli-parent ipc-sub-grid-item ipc-sub-grid-item--span-2'})
```

```python
names = []
ratings = []
year = []
watch_time = []
vote_count = []


for i in soup.find_all('div',{'class':'ipc-poster-card ipc-poster-card--base ipc-poster-card--media-radius ipc-poster-card--dynamic-width sc-d017d8-0 lkttPu gli-parent ipc-sub-grid-item ipc-sub-grid-item--span-2'}):
    try:
        title_tag = i.find('a', class_='ipc-poster-card__title')
        names.append(title_tag.get_text(strip=True))
    except:
        names.append(np.nan)
    try:
        ratings.append(i.find('span',{'class':'ipc-rating-star--rating'}).text)
    except:
        ratings.append(np.nan)
    try:
        vote_count.append(i.find('span',{'class':'ipc-rating-star--voteCount'}).text)
    except:
        vote_count.append(np.nan)
    try:
        year.append(i.find('div',{'class':'sc-d017d8-5 hJgKYC gtc-title-metadata'}).text)
    except:
        year.append(np.nan)
    try:
        watch_time.append(i.find('span',{'class':'gli-metadata-spacer'}).text)
    except:
        watch_time.append(np.nan)
```

```python
df = pd.DataFrame({
    'Name':names,
    'Ratings':ratings,
    'Vote Count': vote_count,
    'Release Year' : year,
    'Watch Time' : watch_time
})
df
```

```python
df.to_csv('Top Movies.csv')
```
